# TensorFlow - Neural Networks

## Model Training

```python
import tensorflow as tf

input_size = 784
hidden1_size = 700
hidden2_size = 600
hidden3_size = 500
hidden4_size = 400
hidden5_size = 300
hidden6_size = 200
hidden7_size = 100
hidden8_size = 50
hidden9_size = 25
output_size = 10


model = tf.keras.Sequential([
    tf.keras.layers.Dense(hidden1_size, activation='relu', input_shape=(input_size,)),
    tf.keras.layers.Dense(hidden2_size, activation='relu'),
    tf.keras.layers.Dense(hidden3_size, activation='relu'),
    tf.keras.layers.Dense(hidden4_size, activation='relu'),
    tf.keras.layers.Dense(hidden5_size, activation='relu'),
    tf.keras.layers.Dense(hidden6_size, activation='relu'),
    tf.keras.layers.Dense(hidden7_size, activation='relu'),
    tf.keras.layers.Dense(hidden8_size, activation='relu'),
    tf.keras.layers.Dense(hidden9_size, activation='relu'),
    tf.keras.layers.Dense(output_size, activation='sigmoid')
])


model.compile(optimizer = 'adam',
              loss = 'sparse_categorical_crossentropy',
              metrics = ['accuracy'] )

history = model.fit(X_train_scaled, y_train, epochs = 20, batch_size = 32, validation_split = 0.2)
```

## Model Prediction

```python
y_pred = model.predict(X_test_scaled)
y_pred_array = np.argmax(y_pred, axis = 1)
```

## Model Evaluation

```python
loss_train, acc_train = model.evaluate(X_train_scaled, y_train)
loss_test, acc_test = model.evaluate(X_test_scaled, y_test)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
```

# Loading Data

```python
data = tf.keras.utils.image_dataset_from_directory(
    'image_data',
    shuffle = True,
    image_size=(28, 28),
    batch_size=32,
    validation_split=False)
```

# Data Classes

```python
classes = data.class_names
classes
```

# Train, Test, Validation Split

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

# Data Visualization

```python
plt.figure(figsize=(10,10))

for image, labels in train.take(1):
    for i in range(9):
        plt.subplot(3,3,i+1)
        plt.imshow(image[i].numpy().astype('uint8'))
        plt.title(classes[labels[i]])
        plt.axis('off')
```

# Convolutional Neural Network

```python
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.models import Sequential

model = Sequential([
    layers.Rescaling(1./255),
    layers.Conv2D(16, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Conv2D(32,3, padding='same',activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Conv2D(64, 3, padding='same', activation='relu'),
    layers.MaxPooling2D(),
    
    layers.Flatten(),
    layers.Dropout(0.2),
    layers.Dense(128),
    layers.Dense(len(classes))              
])
```

```python
model.compile(optimizer='adam', loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True), metrics=['accuracy'])
```

# Model Training

```python
epochs_size = 7
history = model.fit(train, validation_data = val, epochs=epochs_size)
```

# Model Evaluation

```python
epochs_range = range(epochs_size)
plt.figure(figsize=(8,8))
plt.subplot(1,2,1)
plt.plot(epochs_range,history.history['accuracy'],label = 'Training Accuracy')
plt.plot(epochs_range, history.history['val_accuracy'],label = 'Validation Accuracy')
plt.title('Accuracy')

plt.subplot(1,2,2)
plt.plot(epochs_range,history.history['loss'],label = 'Training Loss')
plt.plot(epochs_range, history.history['val_loss'],label = 'Validation Loss')
plt.title('Loss')
```
